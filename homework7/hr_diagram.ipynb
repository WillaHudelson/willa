```python
import numpy as np
import matplotlib.pyplot as plt

star_data = np.loadtxt("../../course_assignments/homework7/star_data.txt",
skiprows=1)

print(star_data.shape) #there are 2642 rows
magnitude = star_data[:, 0]
temperature = star_data[:, 1]
print(magnitude.min()) #the brightest star
print(magnitude.max()) #the dimmest star
print(temperature.min()) #the coldest star
print(temperature.max()) #the hottest star

log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)
plt.figure(figsize=(8, 6))
sc = plt.scatter(log_temperature, log_magnitude, c=log_temperature, cmap='RdYlBu', alpha=0.5)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.scatter(log_temperature[magnitude.argmin()], log_magnitude.min(), color="black", marker="*", s=300, label="Brightest")
plt.scatter(log_temperature[magnitude.argmax()], log_magnitude.max(), color="purple", marker="+", s=300, label="Dimmest")
plt.scatter(log_temperature.min(), log_magnitude[temperature.argmin()], color="orange", marker=".", s=300, label="Coldest")
plt.scatter(log_temperature.max(), log_magnitude[temperature.argmax()], color="green", marker="s", s=300, label="Hottest")
plt.title("Hertzsprung-Russel Diagram")
plt.xlabel("Temperature (log K)")
plt.ylabel("Magnitude")
plt.colorbar(sc, label="Temperature (log K)")
plt.legend()
plt.show()
```
