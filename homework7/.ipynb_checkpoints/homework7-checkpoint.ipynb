```python
import numpy as np
import matplotlib.pyplot as plt

def make_sine_wave(x, A, w):
    return A * np.sin(w * x)

x = np.linspace(0, 2 * np.pi, 1000)
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]
linestyles = ["-", "--", "-."]
fig, axes = plt.subplots(1, 3, figsize=(25, 5))
i = 0
for a, f in zip(amplitude, frequency):
    ls = linestyles[i % len(linestyles)]
    axes[0].plot(x, make_sine_wave(x, a, 1), linestyle=ls, label=f"A: {a}")
    axes[1].plot(x, make_sine_wave(x, 1, f), linestyle=ls, label=f"F: {f}")
    axes[2].plot(x, make_sine_wave(x, a, f), linestyle=ls, label=f"A:{a}, F:{f}")
    i += 1
axes[0].set_title("Varying Amplitude", fontsize=15)
axes[1].set_title("Varying Frequency", fontsize=15)
axes[2].set_title("Varying Amplitude and Frequency", fontsize=15)
for ax in axes:
    ax.set_xlabel("x", fontsize=13)
    ax.set_ylabel("y", fontsize=13)
    ax.legend()
plt.tight_layout()
plt.show()
```
